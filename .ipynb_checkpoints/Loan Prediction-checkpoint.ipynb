```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('trainset.csv')
dataset.describe()
```

```python
dataset = dataset.dropna()
```

```python
dataset.describe()
```

```python
features = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']
X = dataset[features]
Y = dataset.Loan_Status
print(Y)

print(X["Gender"].value_counts())
print(X["Education"].value_counts())

print(X["Married"].value_counts())

print(X["Self_Employed"].value_counts())
```

```python
X['Gender'] = X['Gender'].replace(['Female','Male'],[0,1])
X['Education'] = X['Education'].replace(['Graduate','Not Graduate'],[0,1])
X['Married'] = X['Married'].replace(['Yes','No'],[0,1])
X['Self_Employed'] = X['Self_Employed'].replace(['Yes','No'],[0,1])
```

```python
print(Y)
Y = Y.replace(['Y','N'],[1,0])
```

```python
X[X.isnull().any(axis=1)]
X["Property_Area"].value_counts()
print(X['Dependents'])
```

```python
print(X.dtypes)
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
   
# creating one hot encoder object with categorical feature 0 
# indicating the first column 
columnTransformer = ColumnTransformer([('encoder', 
                                        OneHotEncoder(), 
                                        [2,10])], 
                                      remainder='passthrough') 
  
X = np.array(columnTransformer.fit_transform(X)) 
print(X)
```
